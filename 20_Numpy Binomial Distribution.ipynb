```python
# Difference between noemal and binomial- normal(continues) and binomial(discrete)
# i use 500 data point for binomial and under 100 data point for normal dist.

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.normal(loc=50, scale=5, size=1000), kind = 'kde', label='Normal')
sns.displot(random.binomial(n=100, p=0.5, size=1000), kind = 'kde', label='Binomial')
plt.show()
```
